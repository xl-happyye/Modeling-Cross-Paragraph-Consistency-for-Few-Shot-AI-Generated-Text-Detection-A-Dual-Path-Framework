## Vocabulary richness

```python
import numpy as np
import spacy

# Load model to create nlp object
nlp = spacy.load("en_core_web_sm")
def calculate_brunet_index(text, a=0.165):
    if not isinstance(text, str):
        return np.nan

    text = text.strip()
    if text == "":
        return np.nan

    doc = nlp(text)
    tokens = [
        token.text.lower()
        for token in doc
        if not token.is_space and not token.is_punct
    ]

    N = len(tokens)
    V = len(set(tokens))

    if N < 5 or V < 5:
        return np.nan

    return float(V / (N ** a))
```

## Pos-tags distributions

```python
import spacy
from collections import Counter
import pandas as pd

# Load model
nlp = spacy.load("en_core_web_sm")

def get_pos_tag_counts(text):
    """
    Extract POS tag counts from text
    """
    # Get all POS tags supported by spaCy
    pos_tags = ['ADJ', 'ADP', 'ADV', 'AUX', 'CONJ', 'CCONJ', 'DET', 'INTJ',
                'NOUN', 'NUM', 'PART', 'PRON', 'PROPN', 'PUNCT', 'SCONJ',
                'SYM', 'VERB', 'X', 'SPACE']

    # Initialize Counter
    pos_tag_counts = Counter({tag: 0 for tag in pos_tags})

    # Check if input is valid text
    if not isinstance(text, str):
        # If float (NaN) or other non-string type, return zero count
        return dict(pos_tag_counts)

    # Check if empty string or contains only whitespace
    if not text or not text.strip():
        return dict(pos_tag_counts)

    try:
        doc = nlp(text)

        # Count POS tags for each token
        for token in doc:
            # Ensure tag is in pos_tags, if not use 'X' (other)
            tag = token.pos_ if token.pos_ in pos_tags else 'X'
            pos_tag_counts[tag] += 1

        return dict(pos_tag_counts)

    except Exception as e:
        print(f"Error processing text: {e}")
        return dict(pos_tag_counts)
```

```python
def safe_get_pos_tag_counts(text):
    """
    Safe POS tag counting function, handling non-string and empty values
    """
    # Check if NaN or None
    if pd.isna(text):
        return str({})  # Return string representation of empty dictionary

    # Check if string type
    if not isinstance(text, str):
        # Try to convert to string
        try:
            text = str(text)
        except:
            return str({})  # Return empty dictionary if conversion fails

    # If empty string
    if not text.strip():
        return str({})

    # Call the original POS tag function
    try:
        return str(get_pos_tag_counts(text))
    except Exception as e:
        print(f"Error in get_pos_tag_counts: {e}")
        return str({})

# Modified application section
part = 'intro'
source_col = part + '_text'
feature_names = part + '_pos_tag_counts'
df[feature_names] = df[source_col].progress_apply(
    lambda s: safe_get_pos_tag_counts(s)
)

# body section
part = 'body'
source_col = part + '_text'
feature_names = part + '_pos_tag_counts'
df[feature_names] = df[source_col].progress_apply(
    lambda s: safe_get_pos_tag_counts(s)
)

# conclusion section
part = 'conclusion'
source_col = part + '_text'
feature_names = part + '_pos_tag_counts'
df[feature_names] = df[source_col].progress_apply(
    lambda s: safe_get_pos_tag_counts(s)
)
```

## named-entity distributions

```python
def get_named_entity_counts(text):

    # Initialize a Counter with all possible NER tags
    ner_tags = ['PERSON', 'NORP', 'FAC', 'ORG', 'GPE', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW', 'LANGUAGE', 'DATE', 'TIME', 'PERCENT', 'MONEY', 'QUANTITY', 'ORDINAL', 'CARDINAL']
    ner_counts = Counter({tag: 0 for tag in ner_tags})

    try:
      # Process the text using spaCy
      doc = nlp(text)
      # Iterate through each entity in the document and update the counts of NER tags
      for ent in doc.ents:
          ner_counts[ent.label_] += 1

      # Convert Counter to dictionary
      ner_tag_map = dict(ner_counts)

      return ner_tag_map
    except:
      return dict(ner_counts)
```

## Stopwords distributions

```python
from collections import Counter
import spacy
import pandas as pd

# Ensure model is loaded
nlp = spacy.load("en_core_web_sm")

def get_stopwords_count(text):
    try:
        # First check if input is string
        if pd.isna(text):  # Handle NaN values
            return {}

        if not isinstance(text, str):  # Handle non-string types
            # Try to convert to string
            try:
                text = str(text)
            except:
                return {}

        # Check if empty string
        if not text.strip():
            return {}

        # Process text directly, do not convert to lowercase first
        doc = nlp(text)

        # Initialize Counter
        stopwords_count = Counter()

        # Iterate through tokens in document
        for token in doc:
            # Check if is stopword
            if token.is_stop:
                # Use lowercase form for counting, avoid case-sensitive duplicates
                stopwords_count[token.text.lower()] += 1

        # Sort by word frequency in descending order
        sorted_stopwords = dict(sorted(stopwords_count.items(),
                                     key=lambda item: item[1],
                                     reverse=True))
        return sorted_stopwords

    except Exception as e:
        print(f"Error: {e}")
        return {}
```

```python

part = 'intro'
source_col = part + '_text'
feature_names = part + '_stopwords_count'
df[feature_names] = df[source_col].progress_apply(
    lambda s: str(get_stopwords_count(s))  # Add str()
)

# Second section: Main body
part = 'body'
source_col = part + '_text'
feature_names = part + '_stopwords_count'
df[feature_names] = df[source_col].progress_apply(
    lambda s: str(get_stopwords_count(s))  # Add str()
)

# Third section: Conclusion
part = 'conclusion'
source_col = part + '_text'
feature_names = part + '_stopwords_count'
df[feature_names] = df[source_col].progress_apply(
    lambda s: str(get_stopwords_count(s))  # Add str()
)
```

## sentiment analysis

For sentiment analysis, we divide the text part into sentences. For each sentence, we calculate the polarity score, subjectivity score, and count of positive and negative words in that sentence. While evaluating the whole text part, we averaged over the sentences.

```python
import nltk
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
import pandas as pd

# Download nltk data (required for first run)
nltk.download('punkt', quiet=True)

# Load spaCy model and add TextBlob extension
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')

def get_sentiment_scores(text):
    try:
        # First check if input is NaN or non-string
        if pd.isna(text):  # Handle NaN values
            return [], [], 0, 0

        # Check if string type
        if not isinstance(text, str):
            # Try to convert to string
            try:
                text = str(text)
            except:
                return [], [], 0, 0

        # Now can safely call strip()
        if not text or len(text.strip()) == 0:
            return [], [], 0, 0

        # Tokenize the text into sentences
        sentences = nltk.sent_tokenize(text)

        # Initialize variables
        pos_wc = 0
        neg_wc = 0
        polarity_scores = []
        subjectivity_scores = []

        for sentence in sentences:
            doc = nlp(sentence)

            # Get sentiment scores
            polarity_scores.append(doc._.blob.polarity)
            subjectivity_scores.append(doc._.blob.subjectivity)

            # Analyze sentiment assessment details
            if hasattr(doc._.blob, 'sentiment_assessments'):
                for x in doc._.blob.sentiment_assessments.assessments:
                    if x[1] > 0:
                        pos_wc += len(x[0])
                    elif x[1] < 0:
                        neg_wc += len(x[0])

        return polarity_scores, subjectivity_scores, pos_wc, neg_wc

    except Exception as e:
        print(f"Sentiment analysis error: {e}")
        return [], [], 0, 0
```

```python
part = 'intro'
source_col = part + '_text'
F = ['polarity_scores','subjectivity_scores','pos_wc','neg_wc',]
feature_names = [part+'_'+f for f in F]
df[feature_names] = df[source_col].progress_apply(lambda s: pd.Series(get_sentiment_scores(s)))

part = 'body'
source_col = part + '_text'
F = ['polarity_scores','subjectivity_scores','pos_wc','neg_wc',]
feature_names = [part+'_'+f for f in F]
df[feature_names] = df[source_col].progress_apply(lambda s: pd.Series(get_sentiment_scores(s)))

part = 'conclusion'
source_col = part + '_text'
F = ['polarity_scores','subjectivity_scores','pos_wc','neg_wc',]
feature_names = [part+'_'+f for f in F]
df[feature_names] =  df[source_col].progress_apply(lambda s: pd.Series(get_sentiment_scores(s)))
```

## Formality score

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
pipe = pipeline("text-classification", model="s-nlp/roberta-base-formality-ranker")

def get_formality_scores(text):
  try:
      # Tokenize the text into sentences
      sentences = nltk.sent_tokenize(text)

      # Compute formality for each sentence
      scores = []
      for sentence in sentences:
          result = pipe(sentence)
          if result[0]['label'] == 'informal':
            score = round(1 - result[0]['score'],2)
          else:
            score = round(result[0]['score'],2)
          scores.append(score)
      return scores
  except:
      return []
```

```python
df['intro_formality_score'] = df['intro_text'].progress_apply(get_formality_scores)
df['body_formality_score'] = df['body_text'].progress_apply(get_formality_scores)
df['conclusion_formality_score'] = df['conclusion_text'].progress_apply(get_formality_scores)
```

## Perplexity scores calculations

```python
# =========================================================
# 0. Dependencies
# =========================================================
import math
import re
import numpy as np
import pandas as pd
import torch
from tqdm import tqdm
from transformers import GPT2Tokenizer, GPT2LMHeadModel


# =========================================================
# 1. Load GPT-2
# =========================================================
device = "cuda" if torch.cuda.is_available() else "cpu"

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2").to(device)
model.eval()


# =========================================================
# 2. Sentence splitting
# =========================================================
def smart_sentence_split(text):
    if not text or not isinstance(text, str):
        return []
    sents = re.split(r'(?<=[.!?])\s+', text)
    return [s.strip() for s in sents if s.strip()]


# =========================================================
# 3. Single sentence perplexity (safe)
# =========================================================
def sentence_perplexity(sentence, max_len=512):
    try:
        enc = tokenizer(
            sentence,
            return_tensors="pt",
            truncation=True,
            max_length=max_len
        )

        input_ids = enc["input_ids"].to(device)

        with torch.no_grad():
            outputs = model(input_ids, labels=input_ids)
            loss = outputs.loss

        return math.exp(loss.item())
    except:
        return None


# =========================================================
# 4. Paragraph perplexity (mean + std)
# =========================================================
def segment_perplexity(text, max_sent=10):
    sentences = smart_sentence_split(text)

    if len(sentences) == 0:
        return None, None

    # ⭐ Optional speedup (recommended to keep)
    sentences = sentences[:max_sent]

    ppls = []
    for s in sentences:
        ppl = sentence_perplexity(s)
        if ppl is not None and not math.isnan(ppl):
            ppls.append(ppl)

    if len(ppls) == 0:
        return None, None

    return float(np.mean(ppls)), float(np.std(ppls))


# =========================================================
# 5. Main function (no limit)
# =========================================================
def calculate_perplexity(df):
    df = df.reset_index(drop=True)

    intro_mean, intro_std = [], []
    body_mean, body_std = [], []
    conclusion_mean, conclusion_std = [], []

    for _, row in tqdm(df.iterrows(), total=len(df), desc="Calculate perplexity"):
        im, istd = segment_perplexity(row.get("intro_text", ""))
        bm, bstd = segment_perplexity(row.get("body_text", ""))
        cm, cstd = segment_perplexity(row.get("conclusion_text", ""))

        intro_mean.append(im)
        intro_std.append(istd)
        body_mean.append(bm)
        body_std.append(bstd)
        conclusion_mean.append(cm)
        conclusion_std.append(cstd)

    df["intro_mean"] = intro_mean
    df["intro_std"] = intro_std
    df["body_mean"] = body_mean
    df["body_std"] = body_std
    df["conclusion_mean"] = conclusion_mean
    df["conclusion_std"] = conclusion_std

    return df


# =========================================================
# 6. Example run
# =========================================================
if __name__ == "__main__":
    print("=" * 60)
    print("Perplexity Calculation (FULL DATA, NO LIMIT)")
    print("=" * 60)

    # ⚠️ Replace here with your data reading method
    # df = pd.read_csv("your_file.csv")

    print("Total samples:", len(df))

    df = calculate_perplexity(df)

    print("\n📊 Perplexity statistics:")
    for seg in ["intro", "body", "conclusion"]:
        vals = df[f"{seg}_mean"].dropna()
        print(
            f"{seg.capitalize():12s} | "
            f"mean = {vals.mean():.3f}, std = {vals.std():.3f}"
        )

    print("\nFirst 5 results:")
    print(df[
        [
            "intro_mean",
            "body_mean",
            "conclusion_mean"
        ]
    ].head())

    print("\n✅ Perplexity calculation completed")
```
